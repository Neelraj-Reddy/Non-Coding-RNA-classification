```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import warnings
import re
from Bio import SeqIO

warnings.filterwarnings("ignore")


def preprocessing(finput, foutput):
    alphabet = ("B|D|E|F|H|I|J|K|L|M|N|O|P|Q|R|S|V|W|X|Y|Z")
    with open(foutput, 'a') as file:
        for seq_record in SeqIO.parse(finput, "fasta"):
            description_parts = str(seq_record.description).split()
            name_seq = description_parts[-1] if description_parts else ""
            seq = seq_record.seq
            if re.search(alphabet, str(seq)) is not None:
                print(name_seq)
                print("Removed Sequence")
            else:
                file.write("> " + name_seq)
                file.write("\n")
                file.write(str(seq))
                file.write("\n")
                print(name_seq)
                print("Included Sequence")
    print("Finished")


if __name__ == "__main__":
    # Define input and output filenames
    input_filename = r"E:\4th sem\IOBS-2\project\Non Coding RNA classification\data\test_0"  # Change this to your input file name
    output_filename = "preprocessing_test_output.fasta"  # Change this to your desired output file name
    preprocessing(input_filename, output_filename)
```
